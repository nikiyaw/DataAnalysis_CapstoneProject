The raw_fighter_details.csv file was obtained from Kaggle and it is a list of all fighters that fought from . The focus of this project would only be on female fighters so we will filter out the male fighters and only keep female fighters.

## Data exploration

Let's conduct some preliminary data exploration just to learn more about the dataset.

```python
import pandas as pd
import numpy as np

data = pd.read_csv('raw_fighter_details.csv')
data
```

This is where it gets tricky because because the list does not have a column that identifies whether a fighter is a male or female. The only unique identifier we can use to distinguish them apart is their names. 

Remember that we managed to obtain a list of all female fighters from the other jupyter notebook. We can import it and grab the list to conduct the filtering process. 

```python
fight_stats = pd.read_csv('fight_stats.csv',sep=';',index_col=0)
fight_stats
```

```python
fight_stats.reset_index(inplace=True)
fight_stats = fight_stats.rename(columns = {'index': 'R_fighter'})
fight_stats
```

```python
R_array = fight_stats.R_fighter.unique()
B_array = fight_stats.B_fighter.unique()

fighter_array = np.concatenate((R_array, B_array))
fighter_array_final = np.unique(fighter_array)

fighter_list = fighter_array_final.tolist()
fighter_list
```

Perfect! Now we can filter it through and keep the female fighters. 

```python
fighter = data.isin(fighter_list)
fighter = data[fighter['fighter_name'] == True]
fighter
```

## Data cleaning

Manipulating the datatypes into something that we can work with will make life easier in the future so let's also do that. 

```python
fighter.dtypes
```

```python
import re
from datetime import datetime, timedelta
from datetime import date
```

```python
# Convert height from feet to cm
def format_height(x):
    num = re.sub('[^0-9]', '',str(x))
    num_1 = num[0] + '.' + num[1:]
    num_2 = float(num_1)*0.3048 
    result = round(num_2, 2)
    return result

fighter['Height'] = fighter['Height'].map(format_height)


# Convert weight from pounds to kg
def format_weight(y):
    char = re.sub('[^0-9]', '',str(y))
    num = float(char)*0.4536
    result = round(num, 1)
    return result

fighter['Weight'] = fighter['Weight'].map(format_weight)


# Remove all % in multiple rows
def format_percentage(z):
    num = re.sub('[^0-9]', '',str(z))
    num_1 = float(num)/100
    result = round(num_1, 2)
    return result 

fighter['Str_Acc'] = fighter['Str_Acc'].map(format_percentage)
fighter['Str_Def'] = fighter['Str_Def'].map(format_percentage)
fighter['TD_Acc'] = fighter['TD_Acc'].map(format_percentage)
fighter['TD_Def'] = fighter['TD_Def'].map(format_percentage)


# Extract Age using the DOB
fighter['DOB'] = pd.to_datetime(fighter['DOB'])

def format_dob(dob):
    today = date.today()
    age_in_years = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
    return age_in_years

fighter['DOB'] = fighter['DOB'].map(format_dob)
```

```python
# Rename some columns
fighter.rename(columns = {'DOB': 'Age',
                                'Height': 'Height(cm)',
                                'Weight': 'Weight(kg)',
                                'Str_Acc': 'Str_Acc(%)',
                                'Str_Def': 'Str_Def(%)',
                                'TD_Acc': 'TD_Acc(%)',
                                'TD_Def': 'TD_Def(%)'}, inplace=True)
```
