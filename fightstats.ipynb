# Fight statistics for female fighters
The data.csv file was obtained from Kaggle and it is a list of all fights from . For this project, the focus would be solely on female fighters so we will find a way to filter out the fights that were only participated by female fighters.

## Data exploration
Let's conduct some preliminary data exploration just to learn more about the dataset.

```python
import pandas as pd
import numpy as np

data = pd.read_csv('data.csv')
data
```

```python
list(data.columns)
```

```python
data['weight_class'].unique()
```

This is a very good start! As we can see from the weight_class values, we can filter out the categories and keep only the fights that we are interested in.

```python
data_str = data[data['weight_class'] == 'WomenStrawweight']
data_str
```

```python
data_ban = data[data['weight_class'] == 'WomenBantamweight']
data_ban
```

```python
data_fly = data[data['weight_class'] == 'WomenFlyweight']
data_fly
```

```python
data_fea = data[data['weight_class'] == 'WomenFeatherweight']
data_fea
```

It is always a good practice to check for duplicate rows.

```python
# data[data['weight_class'] == 'WomenStrawweight'].drop_duplicates() -192 rows
# data[data['weight_class'] == 'WomenBantamweightweight'].drop_duplicates() -151 rows
# data[data['weight_class'] == 'WomenFlyweight'].drop_duplicates() -112 rows
# data[data['weight_class'] == 'WomenFeatherweight'].drop_duplicates() -16 rows
```

After checking for duplicates, let's combine all 4 separate dataframes into one.

```python
fight_stats = pd.concat([data_str, data_ban, data_fly, data_fea])
fight_stats
```

Let's also find out how many female fighters were participating in these fights and get a list of the names.

```python
R_array = fight_stats.R_fighter.unique()
B_array = fight_stats.B_fighter.unique()

fighter_array = np.concatenate((R_array, B_array))
fighter_array_final = np.unique(fighter_array)
fighter_array_final

fighter_list = fighter_array_final.tolist()
print('Unique fighter list: ', len(fighter_list))
```
